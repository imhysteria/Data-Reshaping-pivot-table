### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
def calculate_mean_salary_without_highest(employee):
    # Remove the rows with the highest salary in each department
    employee = employee.groupby('Department').apply(lambda x: x[x['Salary'] < x['Salary'].max()])
    
    # Calculate the mean salary for each department
    result = employee.groupby('Department')['Salary'].mean().reset_index()
    
    return result

result = calculate_mean_salary_without_highest(employee)
print(result)
```
